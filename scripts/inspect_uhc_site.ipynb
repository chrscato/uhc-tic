```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# Headless browser setup
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://transparency-in-coverage.uhc.com/")
time.sleep(5)  # Allow initial content to load

# ✅ Uncheck "Index"
try:
    index_cb = driver.find_element(By.CSS_SELECTOR, 'input[type="checkbox"][value="index"]')
    if index_cb.is_selected():
        index_cb.click()
        print("🟡 Unchecked 'Index'")
    else:
        print("ℹ️ 'Index' was already unchecked")
except Exception as e:
    print("⚠️ Could not uncheck 'Index':", e)

# ✅ Check "In Network"
try:
    in_network_cb = driver.find_element(By.CSS_SELECTOR, 'input[type="checkbox"][value="in-network"]')
    if not in_network_cb.is_selected():
        in_network_cb.click()
        print("🟢 Checked 'In Network'")
    else:
        print("ℹ️ 'In Network' was already checked")
except Exception as e:
    print("⚠️ Could not check 'In Network':", e)

# Wait for results to update
time.sleep(7)

# Grab all updated file links
all_links = driver.find_elements(By.TAG_NAME, "a")
in_network_links = []

for link in all_links:
    href = link.get_attribute("href")
    if href and "blob.core.windows.net" in href:
        clean_url = href.split("?")[0]
        in_network_links.append(clean_url)

print(f"✅ Found {len(in_network_links)} in-network file links")
for url in in_network_links[:10]:
    print(url)

# Optionally save to file
with open("uhc_in_network_links.txt", "w") as f:
    for url in in_network_links:
        f.write(url + "\n")

driver.quit()

```
