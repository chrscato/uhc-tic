```python
import pandas as pd
import boto3
import io
from pathlib import Path

# S3 configuration
s3_bucket = "commercial-rates"  # or use os.getenv('S3_BUCKET')
s3_prefix = "tic-mrf"
specific_fact_table = "/consolidated/centene_fidelis/fact_tables/memory_efficient_fact_table_20250802_183825.parquet"      # or use os.getenv('S3_PREFIX')
s3_key = f"{s3_prefix}{specific_fact_table}"  
# Initialize S3 client
s3_client = boto3.client('s3')

try:
    # Get object from S3
    response = s3_client.get_object(Bucket=s3_bucket, Key=s3_key)
    
    # Read parquet directly from memory buffer
    df = pd.read_parquet(io.BytesIO(response['Body'].read()))
    
    # Basic info
    print(f"Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
    print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024 / 1024:.1f} MB")
    print(f"Successfully loaded from s3://{s3_bucket}/{s3_key}")

    # Show first few rows
    df.head()
    
except Exception as e:
    print(f"Error loading from S3: {str(e)}")
```

```python
import ast
import numpy as np
# Cell 2 — Parse the `nppes_addresses` JSON‑like strings
# this creates a column of Python lists of dicts

# First convert numpy array strings to regular list strings
def convert_array_to_list(x):
    if isinstance(x, str):
        # Remove 'array(' prefix and dtype=object) suffix if present
        x = x.replace('array(', '').replace(', dtype=object)', '')
        # Now it should be a regular list string that ast.literal_eval can handle
        return ast.literal_eval(x)
    return x

df['nppes_address_list'] = df['nppes_addresses'].apply(convert_array_to_list)

# check you got lists
df['nppes_address_list'].head()
```

```python
# Cell 3 — Normalize out the first address in each list
# (if you have multiple addresses per provider, you can explode first)
address_df = pd.json_normalize(
    df['nppes_address_list'].apply(lambda lst: lst[0] if lst else {})
)

# prefix the columns so it’s clear where they came from
address_df = address_df.add_prefix('nppes_address_')

# attach back onto your main DataFrame
df = pd.concat([df, address_df], axis=1)

# inspect the new columns
df.filter(like='nppes_address_').head(1)
```

```python
# Cell 4 — Cleanup helper cols
# if you no longer need the raw list:
df = df.drop(columns=['nppes_address_list'])

# and if you want you can also drop the original string field:
# df = df.drop(columns=['nppes_addresses'])

df.head(1)
```

```python
# Cell 6 — Parse `nppes_provider_name` (dict) into parts
# -------------------------------------------------------

# The data is already in dictionary format, so we can normalize it directly
provider_name_df = pd.json_normalize(df['nppes_provider_name'])

# prefix the columns so you know their origin
provider_name_df = provider_name_df.add_prefix('nppes_provider_')

# join back onto the main DataFrame
df = pd.concat([df, provider_name_df], axis=1)

# inspect
df.filter(like='nppes_provider_').head(1)
```

```python
import boto3
import pandas as pd
import io
import botocore

# Config
s3_client = boto3.client('s3')
bucket = 'commercial-rates'
prefix = 'commercial_rates/service_code='

# 🚨 Ensure service_code is string with leading zeros (e.g. '00700')
upload_df['service_code'] = upload_df['service_code'].astype(str).str.zfill(5)

# Group by service_code and upload each
for service_code, group_df in upload_df.groupby('service_code'):
    s3_key = f"{prefix}{service_code}/data.parquet"
    print(f"\n📦 Processing: {s3_key} | {len(group_df)} new rows")

    try:
        # Try to fetch existing file
        response = s3_client.get_object(Bucket=bucket, Key=s3_key)
        existing_df = pd.read_parquet(io.BytesIO(response['Body'].read()))
        print(f"🔄 Found existing file with {len(existing_df)} rows")

        # Match schema
        group_df = group_df[existing_df.columns]
        combined_df = pd.concat([existing_df, group_df], ignore_index=True)

    except botocore.exceptions.ClientError as e:
        if e.response['Error']['Code'] == 'NoSuchKey':
            combined_df = group_df.copy()
            print("🆕 No existing file found. Creating new.")
        else:
            raise e

    # ✅ Safe fallback: convert all object columns to strings
    object_cols = combined_df.select_dtypes(include='object').columns
    combined_df[object_cols] = combined_df[object_cols].astype(str)

    # Re-enforce service_code padding before saving
    if 'service_code' in combined_df.columns:
        combined_df['service_code'] = combined_df['service_code'].astype(str).str.zfill(5)

    # Write to Parquet
    buffer = io.BytesIO()
    combined_df.to_parquet(buffer, index=False)
    buffer.seek(0)

    # Upload
    s3_client.put_object(Bucket=bucket, Key=s3_key, Body=buffer)
    print(f"✅ Uploaded {len(combined_df)} total rows to {s3_key}")
```

```python
import boto3

# Config
s3_client = boto3.client('s3')
bucket = 'commercial-rates'
prefix = 'commercial_rates/service_code='

# List all objects under the prefix
response = s3_client.list_objects_v2(Bucket=bucket, Prefix=prefix)

# Track objects to delete
to_delete = []

if 'Contents' in response:
    for obj in response['Contents']:
        key = obj['Key']
        
        # Extract the service_code from the key path
        if 'service_code=' in key:
            part = key.split('service_code=')[-1].split('/')[0]
            
            # Delete if length of service_code is less than 5
            if len(part) < 5:
                print(f"🗑️ Marked for deletion: {key}")
                to_delete.append({'Key': key})

# Perform deletion (in batches of 1000 max per request)
if to_delete:
    response = s3_client.delete_objects(
        Bucket=bucket,
        Delete={'Objects': to_delete}
    )
    print(f"\n✅ Deleted {len(to_delete)} malformed service_code objects.")
else:
    print("✅ No malformed service_code folders found.")
```
