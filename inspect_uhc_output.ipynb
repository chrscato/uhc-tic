```python
import numpy as np
addresses_df = joined_df['addresses'].apply(
    lambda x: x[0] if isinstance(x, (list, np.ndarray)) and len(x) > 0 else {}
).apply(pd.Series)

```

```python
print("Extracted columns:", addresses_df.columns.tolist())
```

```python
addresses_df = addresses_df.rename(columns={'zip': 'full_zip'})

joined_df = pd.concat([joined_df, addresses_df], axis=1)
display(joined_df.head(1))
```

```python
for i, row in joined_df['addresses'].dropna().head(10).items():
    print(f"Row {i}:")
    print("  Type:", type(row))
    print("  Value:", row)
    if isinstance(row, list) and len(row) > 0:
        print("  First dict keys:", row[0].keys())
        print("  ZIP:", row[0].get('zip'))
    else:
        print("  Skipped - not a list or empty.")
    print("-" * 40)
```

```python
joined_df['zip5'] = joined_df['full_zip'].astype(str).str[:5]

```

```python
# replace the path with wherever your .xlsx lives
cbsa_xwalk = pd.read_excel(r"scripts\cbsa_codes.xlsx", 
                            # or whatever sheet you need
                          engine='openpyxl')     # necessary if you haven’t installed xlrd >2.0

# inspect the first few rows
cbsa_xwalk.head()
```

merge the two cbsa files

```python
merged_cbsa = pd.merge(zip_xwalk, cbsa_xwalk, left_on='CBSA', right_on='CBSA Code', how='inner')
merged_cbsa.head()
```

join the cbsa to fact table

```python
fact_table = pd.merge(
    joined_df,
    merged_cbsa,
    left_on='zip5',
    right_on='ZIP',
    how='left',
    suffixes=('_l','_r')
)
fact_table.head(1)
```

```python
target_tins = values_to_check # replace with your actual values
ga_fact_table_tins = fact_table[fact_table['tin_value'].isin(target_tins)]
```

```python
# Remove duplicate columns by keeping the first occurrence
ga_fact_table_tins = ga_fact_table_tins.loc[:, ~ga_fact_table_tins.columns.duplicated()]
```

```python
ga_fact_table_tins_cleaned = ga_fact_table_tins[ga_fact_table_tins['state'] == 'GA']
display(ga_fact_table_tins_cleaned.head(1))
print("unique CBSA Title: ", ga_fact_table_tins_cleaned['CBSA Title'].unique())
```

```python
for col in ga_fact_table_tins_cleaned.columns:
    if pd.api.types.is_datetime64tz_dtype(ga_fact_table_tins_cleaned[col]):
        print(f"Converting timezone-aware column: {col}")
        ga_fact_table_tins_cleaned[col] = ga_fact_table_tins_cleaned[col].dt.tz_localize(None)


ga_fact_table_tins_cleaned.to_csv("GA_UHC.csv", index=True)
```
